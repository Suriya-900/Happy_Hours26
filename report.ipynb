```python
import pandas as pd

df = pd.read_csv("tracker_data.csv")

df["Date"] = pd.to_datetime(df["Date"])

# Monthly summary
df["Month"] = df["Date"].dt.to_period("M")

monthly = df.groupby(["Phone","Month"]).agg({
    "Daily_Weight_Loss":"sum",
    "Calories_Intake":"mean"
}).reset_index()

monthly.head()
```
